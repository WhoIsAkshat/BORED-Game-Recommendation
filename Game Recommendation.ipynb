```python
def convert_reviews(text):
    ans=[]
    text= text.split(",")
    ans.append(text[0])
    for i in text:
        if '%' in i:
            ans.append(i[2:5])
    return ans
            
```

```python
def convert_price(text):
    l=[]
    if '$' in text:
        l.append(math.ceil(eval(text.replace('$','',1))))
    else:
        l.append(0)
    return l
```

```python
games['original_price'] = games['original_price'].apply(convert_price)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_g['tags'] = new_g['tags'].apply(stem)

#After this i repeated Vectorization
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(game):
    game_idx= int(new_g[new_g['name'] == game].new_index)-1
    distances=similarity[game_idx]
    game_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    for i in game_list:
        print(new_g.name[i[0]])
```
